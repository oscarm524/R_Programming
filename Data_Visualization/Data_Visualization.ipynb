# My First Plot

```r
## Defining the values of x
x = seq(1, 7, len = 100)

## Defining y1 and y2
y1 = x^2 + 2*x + 3
y2 = 3*x + 4

## Defining the pdf file to save plot
# pdf('my_first_plot.pdf', width = 10, height = 8)

## Plotting x against y1 and y2
plot(x, y1, type = 'l', lwd = 2, col = 'blue', ylab = 'Y values', xlab = 'X values')
lines(x, y2, lwd = 2, col = 'orange')

## Adding grid
grid()

# dev.off()
```

# Histogram

```r
## Reading titanic datafile
titanic = read.csv(file = 'titanic.csv', header = TRUE)
head(titanic)
```

```r
## Defining the pdf file to save plot
# pdf('histogram.pdf', width = 10, height = 8)

## Creating a histogram of passenger ages
hist(titanic$Age, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

# dev.off()
```

# Subplots of Histograms

```r
## Defining the pdf file to save plot
# pdf('histogram_subplots.pdf', width = 14, height = 8)

## Creating a figure with 4 histograms
par(mfrow = c(2, 2))

## Plotting the first histogram
hist(titanic$Age, breaks = 3, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

## Plotting the second histogram
hist(titanic$Age, breaks = 10, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

## Plotting the third histogram
hist(titanic$Age, breaks = 20, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

## Plotting the fourth histogram
hist(titanic$Age, breaks = 50, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

# dev.off()
```

# Bar Chart

```r
## Defining the pdf file to save plot
# pdf('bar_chart.pdf', width = 10, height = 8)

## Creating a bar chart of passenger gender
barplot(table(titanic$Sex), col = 'deepskyblue3', xlab = 'Passenger Age', ylab = 'Count', ylim = c(0, 600))
box()

# dev.off()
```

# Grouped Bar Chart

```r
## Defining the pdf file to save plot
# pdf('group_bar_chart.pdf', width = 10, height = 8)

## Creating a grouped bar chart for survived and sex
barplot(table(titanic$Sex, titanic$Survived), beside = TRUE, col = c('darkorange2', 'deepskyblue3'), legend = c('women', 'men'), xlab = 'Survived Status', ylim = c(0, 600), main = 'Survived Status by Gender') 
box()

# dev.off()
```

# Stacked Bar Chart

```r
## Defining the pdf file to save plot
# pdf('stacked_bar_chart.pdf', width = 10, height = 8)

## Creating a stacked bar chart for survived and sex
barplot(table(titanic$Sex, titanic$Survived), col = c('darkorange2', 'deepskyblue3'), legend = c('women', 'men'), xlab = 'Survived Status', ylim = c(0, 600), main = 'Survived Status by Gender') 
box()

# dev.off()
```

# Stacked Percentage Bar Chart

```r
## Defining the pdf file to save plot
# pdf('stacked_percentage_bar_chart.pdf', width = 10, height = 8)

## Creating a stacked percentage bar chart for survived and sex
barplot(100*prop.table(table(titanic$Sex, titanic$Survived), 2), col = c('darkorange2', 'deepskyblue3'), legend = c('women', 'men'), xlab = 'Survived Status', ylim = c(0, 120), main = 'Survived Status by Gender') 
box()

# dev.off()
```

# Pie Chart

```r
## Defining the pdf file to save plot
# pdf('pie_chart.pdf', width = 10, height = 8)

## Creating a pie chart of Sex
pie(table(titanic$Sex), col = c('darkorange2', 'deepskyblue3'), labels = c('Female - 35%', 'Male - 65%'))

# dev.off()
```

# Box Plot

```r
## Defining the pdf file to save plot
# pdf('boxplot.pdf', width = 10, height = 8)

## Creating a box plot of Age
boxplot(titanic$Age, xlab = 'Age', col = 'deepskyblue3')
grid()

# dev.off()
```

# Side-by-Side Box Plot

```r
## Defining the pdf file to save plot
# pdf('side_by_side_boxplot.pdf', width = 12, height = 8)

## Creating a side-by-side box plot 
boxplot(Age ~ Survived, data = titanic, col = c('darkorange2', 'deepskyblue3'), xaxt = 'n', xlab = '')
axis(side = 1, at = 1:2, labels = c('Not Survived', 'Survived'))
grid()

# dev.off()
```

# Scatter Plot

```r
## Defining the pdf file to save plot
# pdf('scatter_plot.pdf', width = 10, height = 8)

## Creating scatter-plot between age and fare
plot(titanic$Age, titanic$Fare, col = 'deepskyblue3', pch = 16, xlab = 'Age', ylab = 'Fare')
grid()

# dev.off()
```
