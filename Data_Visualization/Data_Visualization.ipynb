# My First Plot

```r
## Defining the values of x
x = seq(1, 7, len = 100)

## Defining y1 and y2
y1 = x^2 + 2*x + 3
y2 = 3*x + 4

## Defining the pdf file to save plot
# pdf('my_first_plot.pdf', width = 10, height = 8)

## Plotting x against y1 and y2
plot(x, y1, type = 'l', lwd = 2, col = 'blue', ylab = 'Y values', xlab = 'X values')
lines(x, y2, lwd = 2, col = 'orange')

## Adding grid
grid()

# dev.off()
```

# Histogram

```r
## Reading titanic datafile
titanic = read.csv(file = 'titanic.csv', header = TRUE)
head(titanic)
```

```r
## Defining the pdf file to save plot
# pdf('histogram.pdf', width = 10, height = 8)

## Creating a histogram of passenger ages
hist(titanic$Age, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

# dev.off()
```

# Subplots of Histograms

```r
## Defining the pdf file to save plot
# pdf('histogram_subplots.pdf', width = 14, height = 8)

## Creating a figure with 4 histograms
par(mfrow = c(2, 2))

## Plotting the first histogram
hist(titanic$Age, breaks = 3, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

## Plotting the second histogram
hist(titanic$Age, breaks = 10, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

## Plotting the third histogram
hist(titanic$Age, breaks = 20, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

## Plotting the fourth histogram
hist(titanic$Age, breaks = 50, xlab = 'Passenger Age', ylab = 'Count', main = 'Histogram of Passenger Age', col = 'deepskyblue3')
box()

# dev.off()
```
